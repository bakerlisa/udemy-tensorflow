## Export pandas DataFrames to CSV or excel files

```python
file_url = '/content/drive/MyDrive/Colab Notebooks/coding_dojo/files/mortgages.csv'
df = pd.read_csv(file_url)
df.head()
```

```python
mortgage_filter = df['Mortgage Name'] == '30 Year'
interest_filter = df['Interest Rate'] == 0.03
df = df.loc[interest_filter & mortgage_filter, :]
```

```python
df.head()
```

```python
# Exporting the data - 2 ways
df.to_csv(path_or_buf="OneMortgage.csv", index=False)
```
